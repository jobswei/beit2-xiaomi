```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_value_lis(name:str, file):
    lis=[]
    with open(file,"r") as fp:
        txt=fp.readlines()
    for line in txt:
        try:
            lis.append(eval(line)[name])
        except:
            continue
    return lis
```

```python
file="/home/xiaomi/unilm/beit2/beit_output/log.txt"
train_loss=get_value_lis("train_loss",file)
test_loss=get_value_lis("test_loss",file)
plt.plot(train_loss)
plt.plot(test_loss)
plt.legend(["train_loss","test_loss"])
```

```python
file="/home/xiaomi/unilm/beit2/outputs/log.txt"
train_loss=get_value_lis("train_total_loss",file)
test_loss=get_value_lis("test_total_loss",file)
plt.plot(train_loss)
plt.plot(test_loss)
plt.legend(["train_total_loss","test_total_loss"])
```
